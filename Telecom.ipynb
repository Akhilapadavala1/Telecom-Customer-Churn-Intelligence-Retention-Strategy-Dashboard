```python
high_risk = df[
    (df['Churn_Label']==1)
    &
    (df['Tenure_in_Months']<12)
]

high_risk.head() #High Risk Customers
```

```python
def risk_score(row):

    if row['Churn_Label'] == 1:
        return "High Risk"

    elif row['Tenure_in_Months'] < 12:
        return "Medium Risk"

    else:
        return "Low Risk"

df['Risk_Score'] = df.apply(
    risk_score,
    axis=1
) #Create Risk Score (Project Requirement)
```

```python
df['Heavy_Usage'] = df['Avg_Monthly_GB_Download'].apply(
    lambda x: 'Yes' if x > 100 else 'No'
)
```

```python
def customer_risk(row):

    if row['Churn_Label'] == 1:
        return "High Risk"

    elif row['Tenure_in_Months'] < 12:
        return "Medium Risk"

    else:
        return "Low Risk"

df['Risk_Score'] = df.apply(
    customer_risk,
    axis=1
)

df[['Customer_ID','Risk_Score']].head() #Create Customer Risk Score (Function + If-Else)
```

```python
df['Heavy_Usage'] = df['Avg_Monthly_GB_Download'].apply(
    lambda x: 'Yes' if x > 100 else 'No'
)

df[['Avg_Monthly_GB_Download','Heavy_Usage']].head() #Use Lambda for Heavy Usage Flag
```
